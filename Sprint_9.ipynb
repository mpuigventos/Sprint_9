```python
def conversor_temperatura(temperatura, unitat):
    """
    Introduim una temperatura Celsius i la convertim a Kelvin i Fahrenheit.
    """
    while unitat != "Celsius" or unitat != "celsius" or unitat != "Kelvin" or unitat != "kelvin" or unitat != "Fahrenheit" or unitat != "fahrenheit":
        print("El formato no es correcto")
        temperatura = float(input("Introduce la temperatura en grados Celisus que quieres convertir:"))
        unitat = input("Introduce la unidad: Celsius, Fahrenheit o Kelvin:")
        if unitat == "Celsius" or unitat == "celsius" or unitat == "Kelvin" or unitat == "kelvin" or unitat == "Fahrenheit" or unitat == "fahrenheit":
            break
    conversor = {}

    if unitat == "Celsius" or unitat == "celsius":
        celsius = temperatura
        kelvin = celsius + 273.15
        fahrenheit = (celsius*9/5)+32
        conversor = {"Celsius": celsius,"Kelvin": kelvin, "Fahrenheit":fahrenheit}
    elif unitat == "Kelvin" or unitat == "kelvin":
        kelvin = temperatura
        celsius = kelvin - 273,15
        fahrenheit = (celsius*9/5)+32
        conversor = {"Celsius": celsius,"Kelvin": kelvin, "Fahrenheit":fahrenheit}
    elif unitat == "Fahrenheit" or unitat == "fahrenheit":
        fahrenheit = temperatura
        celsius = ((fahrenheit-32)/9)*5
        kelvin = celsius + 273.15
        conversor = {"Celsius": celsius,"Kelvin": kelvin, "Fahrenheit":fahrenheit}
    return conversor

entrada_temperatura = float(input("Introduce la temperatura que quieres convertir:"))
entrada_unitat = input("Introduce la unidad: Celsius, Fahrenheit o Kelvin:")
temperatures = conversor_temperatura(entrada_temperatura, entrada_unitat)
temperatures

```

# 3. Comptador de paraules d'un text.

Escriu una funci√≥ que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casu√≠stiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina √©s la llargada mitjana de les paraules del text que has escrit? ‚ÄúHola com va?‚Äù hauria de retornar (4+3+2) / 3 = 3

```python
import string
def comptador_paraules(frase):
    def limpiar_frase(texto):
        minus = texto.lower() #convertimos a minuscula la frase
        simbolos = str.maketrans("","",string.punctuation) # eliminamos los simbolos de la frase
        sense_signes = minus.translate(simbolos)
        paraules_netes = sense_signes.split(" ") # separamos palabras por los espacios
        return paraules_netes
    paraules = limpiar_frase(frase)
    comptador = 0
    dic = dict()
    llargada = []
    llargada_paraules = 0
    mitjana_llargada_paraules = 0
    for paraula in paraules: # bucle for para obtener la media de letras por palabra
        llargada.append(len(paraula))
        llargada_paraules = llargada_paraules+len(paraula)
    mitjana_llargada_paraules = round(llargada_paraules/len(llargada),2)
    for paraula in paraules: # buble para obtener el contador de palabras
        if paraula in dic:
            comptador +=1
            dic.update({paraula: comptador})
        else:
            comptador = 1
            dic.update({paraula: comptador})
    return dic, f"la media de letras por palabra es: {mitjana_llargada_paraules}"

introdueix_frase = input("Escriu la frase de la que vulguis comptar les paraules")
comptador_paraules(introdueix_frase)


```

## 4. Diccionari invers (amb possibilitat de duplicats)

Resulta que el client t√© una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al rev√©s, √©s a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original s√≥n √∫nics; si aquest no √©s el cas, la funci√≥ hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

Fixa‚Äôt en el seg√ºents exemples per veure l‚Äôoutput esperat.

```python
def girar_diccionari(diccionari):
    rev_dic = dict()
    for x,y in diccionari.items(): # bucle para poner intercambiar claves por valores y en el inverso poner los nuevos valores en listas
        if y in rev_dic.keys():
            rev_dic[y].append(x)
    
        else:
            rev_dic[y] = [x]
    for x,y in rev_dic.items(): #bucle per treure la llista dels que nomes tenen un element
        if len(y) == 1:
            rev_dic[x] = y[0]
    for clau,valor in rev_dic.items():
        if type(valor) is list:
            print(f"Aquests valors {valor} tenen la mateixa clau: {clau}")
    return rev_dic

diccio = {"a":1,"b":2, "c":2,"d":3,"f":1,"g":1,"h":2, "l":2}
reverse_dictionary=girar_diccionari(diccio)
reverse_dictionary

```

## 1. Comptador i endre√ßador de paraules d'un text.

El client va quedar content amb el comptador de paraules, per√≤ ara vol llegir arxius TXT i que calculi la freq√º√®ncia de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, √©s a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
import string
import os
def calaix_paraules(ruta):
    #c√†rrega del fitxer i inclusio a una variable
    texto = open(ruta, encoding='utf-8')
    text_str = ""
    for linea in texto:
        text_str+=linea
    #neteja fitxer
    minus = text_str.lower()
    simbolos = str.maketrans("","",string.punctuation+'¬°¬ø') # eliminamos los simbolos de la frase
    sense_signes = minus.translate(simbolos)
    limpieza = sense_signes.replace("\n"," ")
    text = limpieza.split()

    # Generar diccionari per les lletres i per les paraules
    lletres=dict()
    diccio_paraules = dict()

    for paraula in text: #Bucle para obtener diccionario con la primera letra de cada palabra
        if paraula[0] not in lletres:
            lletres.update({paraula[0]: 0})

    for paraula in text: # Bucle para recorrer el texto, y a√±adir las palabras como diccioanrio
        inicial = paraula[0]
        if lletres[inicial] == 0: #Si la incial tiene valor 0, sustituimos por la palbara y ponemos el contador en 1
            lletres[inicial] = {paraula: 1}
        else:  #si no es 0, intorducimos la palabra e incrementamos el contador
            diccio_paraules = lletres[inicial]
            if paraula in diccio_paraules:
                diccio_paraules[paraula] += 1
            else:
                diccio_paraules[paraula] = 1
    lletres_ordenades = dict(sorted(lletres.items(),reverse=False))
    return lletres_ordenades

fitxer= "tu me quieres blanca.txt"
ruta = os.path.abspath(fitxer)
calaix_paraules(ruta)
```

## 1. Generador de Contrasenyes

Explora el funcionament del m√≤dul random de la llibreria numpy. üîó Random module in NumPy

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d‚Äôaniversari o similars. Per a solucionar-ho ens ha encarregat una funci√≥ de Python que generi contrasenyes m√©s segures. La funci√≥ ha de dependre dels seg√ºents par√†metres:

longitud (int): Longitud de la contrasenya
majuscules (bool = True): Si hi ha d‚Äôapar√®ixer maj√∫scules
minuscules (bool = True): Si hi ha d‚Äôapar√®ixer min√∫scules
numeros (bool = True): Si hi ha d‚Äôapar√®ixer n√∫meros
signes (bool = False): Si hi ha d‚Äôapar√®ixer car√†cters especials (,-$? o similars)
Aix√≠ doncs, si executem la funci√≥ de la seg√ºent manera:

crear_contrasenya(10, True, True, True, True)

Haur√≠em d‚Äôobtenir un output (que hem de poder guardar) de l‚Äôestil:

9Er,5Vn8P$

Assegura‚Äôt que es compleixin tots els criteris, i que aquestes contrasenyes siguin realment aleat√≤ries.

(EXTRA) Explora com podr√≠em fer que la funci√≥ copi√©s la contrasenya autom√†ticament al porta papers de l‚Äôordinador (com si l‚Äôhagu√©ssim seleccionat i fet ctrl+copy).

```python
from numpy import random
import string
import pyperclip as clip

def crear_contrasenya(long,numeros,mayuscula,minuscula,caracter):
    #lista para introducir todos las opciones de caracteres
    caracteres_totales = []
    #lista para asegurarnos que aparecen todas las opciones
    garantizados = []

    #condicion para que introduzca en las listas los caracteres en cuestin si se escoge la opcion True
    if numeros:
        caracteres_totales.append(string.digits)
        garantizados.append(random.choice(list(string.digits)))
    if mayuscula:
        caracteres_totales.append(string.ascii_uppercase)
        garantizados.append(random.choice(list(string.ascii_uppercase)))
    if minuscula:
        caracteres_totales.append(string.ascii_lowercase)
        garantizados.append(random.choice(list(string.ascii_lowercase))) 
    if caracter:
        caracteres_totales.append(string.punctuation)
        garantizados.append(random.choice(list(string.punctuation)))

    # pasar a cadena string la lista del total de caracteres
    lista = "".join(caracteres_totales)
    # calcular caracteres que faltan para llegar a la longitud indicada
    caracteres_restantes = long - len(garantizados)

    restantes_aleatorios = [random.choice(list(lista)) for i in range(caracteres_restantes)]
    # Combinar y mezclar
    password_list = garantizados + restantes_aleatorios
    random.shuffle(password_list)

    new_password=""
    for i in range(long):
        new_password = new_password+random.choice(password_list)
    clip.copy(new_password)
    print("Tens la contrassenya copiada.")
    #imprimir qu ela contrassenya esta copiada
    return new_password


password = crear_contrasenya(10,True,True,True,True)

clip.paste()

```

```python
import os
import pandas as pd

def dades_futbol(ruta_fitxer):
    historics = open(ruta_fitxer, encoding='utf-8')
    historics_str = ""
    #print(str(texto.read()))
    #print(type(texto))
    for linea in historics:
        historics_str += linea
        
    #separem per tenir els resultats de cada partit
    limpieza_resultados = historics_str.replace("\t",",").replace("\n",",").replace("-",",")
    #resultats = limpieza_resultados.split("\n")
    #creem subllista per equips de casa, marcador i equip visitant
    resultat = limpieza_resultados.split(",")
    equip_casa = resultat[::4]
    gols_casa = resultat[1::4]
    gols_visitant = resultat[2::4]
    equip_visitant = resultat[3::4]

    #convertir a int les llistes de gols
    for gol in range(len(gols_casa)):
        gols_casa[gol] = int(gols_casa[gol])
    for gol in range(len(gols_visitant)):
        gols_visitant[gol] = int(gols_visitant[gol])

    # generar lista per cada dada
    llista_partits= list(zip(equip_casa,gols_casa,gols_visitant,equip_visitant))


    #Generar llistat d'equips
    llistatEquips = []
    for equip in equip_casa:
        if equip not in llistatEquips and equip != "":
            llistatEquips.append(equip)
    for equip in equip_visitant:
        if equip not in llistatEquips and equip != "":
            llistatEquips.append(equip)

    #crear DataFrame amb dades columnes necessaries i valors = 0
    classificacio = pd.DataFrame(index=llistatEquips, columns=["Partits guanyats", "Partits empatats","Partits perduts", "Gols a favor","Gols en contra", "Punts"])  
    for columna in classificacio.columns:
        classificacio[columna] = classificacio[columna].fillna(0)

    #carregar dades Dataframe
    for partit in llista_partits:
        if partit[1]<partit[2]:
            classificacio.at[partit[0], "Partits perduts"] = classificacio.at[partit[0], "Partits perduts"] + 1
            classificacio.at[partit[3], "Partits guanyats"] = classificacio.at[partit[3], "Partits guanyats"] + 1
            classificacio.at[partit[3], "Punts"] = classificacio.at[partit[3], "Punts"] + 3
            classificacio.at[partit[0], "Gols a favor"] = classificacio.at[partit[0], "Gols a favor"] + partit[1]
            classificacio.at[partit[0], "Gols en contra"] = classificacio.at[partit[0], "Gols en contra"] + partit[2]
            classificacio.at[partit[3], "Gols a favor"] = classificacio.at[partit[3], "Gols a favor"] + partit[2]
            classificacio.at[partit[3], "Gols en contra"] = classificacio.at[partit[3], "Gols en contra"] + partit[1]

        elif partit[1]>partit[2]:
            classificacio.at[partit[0], "Partits guanyats"] = classificacio.at[partit[0], "Partits guanyats"] + 1
            classificacio.at[partit[3], "Partits guanyats"] = classificacio.at[partit[3], "Partits guanyats"] + 1
            classificacio.at[partit[0], "Punts"] = classificacio.at[partit[0], "Punts"] + 3
            classificacio.at[partit[0], "Gols a favor"] = classificacio.at[partit[0], "Gols a favor"] + partit[1]
            classificacio.at[partit[0], "Gols en contra"] = classificacio.at[partit[0], "Gols en contra"] + partit[2]
            classificacio.at[partit[3], "Gols a favor"] = classificacio.at[partit[3], "Gols a favor"] + partit[2]
            classificacio.at[partit[3], "Gols en contra"] = classificacio.at[partit[3], "Gols en contra"] + partit[1]
        elif partit[1]==partit[2]:
            classificacio.at[partit[0], "Partits empatats"] = classificacio.at[partit[0], "Partits empatats"] + 1
            classificacio.at[partit[3], "Partits empatats"] = classificacio.at[partit[3], "Partits empatats"] + 1
            classificacio.at[partit[0], "Punts"] = classificacio.at[partit[0], "Punts"] + 1
            classificacio.at[partit[3], "Punts"] = classificacio.at[partit[3], "Punts"] + 1
            classificacio.at[partit[0], "Gols a favor"] = classificacio.at[partit[0], "Gols a favor"] + partit[1]
            classificacio.at[partit[0], "Gols en contra"] = classificacio.at[partit[0], "Gols en contra"] + partit[2]
            classificacio.at[partit[3], "Gols a favor"] = classificacio.at[partit[3], "Gols a favor"] + partit[2]
            classificacio.at[partit[3], "Gols en contra"] = classificacio.at[partit[3], "Gols en contra"] + partit[1]

    gols_equip = classificacio.set_index(classificacio.index)['Gols a favor'].to_dict()
    equip_golejador= classificacio['Gols a favor'].idxmax()
    equip_golejat = classificacio['Gols en contra'].idxmax()
    print(f"Llistat de gols per equip: \n{gols_equip} \n\n L'equip m√©s golejador: \n{equip_golejador} \n\n L'equip m√©s golejat: \n{equip_golejat}")
    return gols_equip, equip_golejador, equip_golejat, classificacio

fitxer = "historic partits.txt"
ruta= os.path.abspath(fitxer)
dades_futbol(ruta)
```
